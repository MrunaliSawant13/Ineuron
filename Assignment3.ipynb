3. What statement creates a function?

ANS: The 'def' keyword following the 'function name' will create a function. 
     For example:
         def new_function():
            print("Hello World")
        This will create a function named 'new_function'

ANS: When the function call returns, the local scope variables are forgotten and thus the local scope is destroyed.
